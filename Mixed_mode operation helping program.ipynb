```python
import numpy as np
import pandas as pd 
import math as m 
import pythermalcomfort as ptc
```

```python
def convert_mon_to_hour (mon):
  if mon == 1:
    add_mon_hour = 0
  elif mon == 2:
    add_mon_hour = 0+(31*24)
  elif mon ==3:
    add_mon_hour = (31*24)+(28*24)
  elif mon == 4:
    add_mon_hour = 2*(31*24)+(28*24)
  elif mon == 5:
    add_mon_hour = 2*(31*24)+(28*24)+(30*24)
  elif mon == 6:
    add_mon_hour = 3*(31*24)+(28*24)+(30*24)
  elif mon == 7:
    add_mon_hour = 3*(31*24)+(28*24)+ 2*(30*24)
  elif mon == 8:
    add_mon_hour = 4*(31*24)+(28*24)+ 2*(30*24)
  elif mon == 9:
    add_mon_hour = 5*(31*24)+(28*24)+ 2*(30*24)
  elif mon == 10:
    add_mon_hour = 5*(31*24)+(28*24)+ 3*(30*24)
  elif mon==11:
    add_mon_hour = 6*(31*24)+(28*24)+ 3*(30*24)
  elif mon==12:
    add_mon_hour = 6*(31*24)+(28*24)+ 4*(30*24)
  else:
    pass
  return (add_mon_hour)
```

```python
def convert_date_to_hour (day):
  if day>=1 and day<=31:
    #counting all the hours that have passed till now
    previous_days = day-1
    add_day_hours = previous_days*24
  else:
    pass
  return (add_day_hours)
```

```python
def mixed_mode (place, time):
  if place==1:
    dataset = pd.read_csv("kolkata.csv")
  elif place ==2:
    dataset = pd.read_csv("bhubaneshwar.csv")
  elif place ==3:
    dataset = pd.read_csv("vishakhapatnam.csv")
  elif place ==4:
    dataset = pd.read_csv("hyderabad.csv")
  elif place ==5:
    dataset = pd.read_csv("chennai.csv")
  elif place ==6:
    dataset = pd.read_csv("pune.csv")
  elif place ==7:
    dataset = pd.read_csv("surat.csv")
  elif place ==8:
    dataset = pd.read_csv("ahmedabad.csv")
  else:
    pass
  t_opi = dataset.iloc[:,4].values # operative indoor temp
  hour = dataset.iloc[:,0].values 
  dbt_out = dataset.iloc[:,1].values
  dbt_in = dataset.iloc[:,3].values 
  t_mrt = dataset.iloc[:,2].values  # mean radiant temp
  t_rmt = dataset.iloc[:,5].values # running mean outdoor temp
  vel = dataset.iloc[:,6].values 
  i = time
  from pythermalcomfort.models import adaptive_ashrae
  result_ashrae55 = adaptive_ashrae(tdb=dbt_in[i], tr=t_mrt[i],t_running_mean=t_rmt[i],v=vel[i],units='SI')
  if result_ashrae55['acceptability_80']==True :
    if dbt_in[i] > dbt_out[i] :
      window = "Open"
      ac = "Off"
      heater = "Off"
    else:
      window = "Closed"
      ac = "Off"
      heater = "Off"
  elif result_ashrae55['tmp_cmf_80_up']<t_opi[i]:
      window = "Closed"
      ac = "On"
      heater = "Off"
  elif result_ashrae55['tmp_cmf_80_low']>t_opi[i]:
      window = "Closed"
      ac = "Off"
      heater = "On"
  return (window, ac, heater) 
```

```python
print("\nMixed-mode operation assistance provider \n")
city = int(input("Select the city where you are located by entering the number mentioned in front of it:\n 1.) Kolkata\n 2.) Bhubaneshwar\n 3.) Vishakhapatnam\n 4.) Hyderabad\n 5.) Chennai\n 6.) Pune\n 7.) Surat\n 8.) Ahmedabad\n "))
month = int(input("\nNow provide date, first by selecting the appropriate month:\n 1.) January\n 2.) February\n 3.) March \n 4.) April\n 5.) May\n 6.) June\n 7.) July\n 8.) August\n 9.)Septemper\n 10.) October\n 11.) November\n 12.) December\n"))
date = int(input("\nEnter date of the month \n"))
hour = float(input("\nEnter the time approximated to nearest hour mark for which you require information in 24-hour format {Eg. 6PM = 18 hrs 10AM = 10 hrs} \n"))
month_hour = convert_mon_to_hour(month)
date_hour = convert_date_to_hour(date)
annual_hour = int(month_hour+date_hour+hour)
openings, hvac , heating = mixed_mode(city, annual_hour)
print("\nThe annual hour {Starting from 00:00 AM on 1st January} provided by the user date and time is = ", annual_hour,"\n")
print("\nBased on our simulation results analysis, For the selected city and the time provided by the user. \nAt this hour \nUser should keep the windows: ",openings ,"\nAir-Conditioning System should be turned: ",hvac,"\nHeater should be turned: ", heating)
```
